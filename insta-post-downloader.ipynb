```python
import instaloader
from pathlib import Path

enter_username = input("Enter username: ")   # target profile
DOWNLOAD_DIR = Path("/Users/satyamgajjar/Downloads") / f"{enter_username}"

# enable video downloads
L = instaloader.Instaloader(download_videos=True)

# Make folder if missing
DOWNLOAD_DIR.mkdir(parents=True, exist_ok=True)

profile = instaloader.Profile.from_username(L.context, enter_username)
print(f"Profile: {profile.username} | Posts: {profile.mediacount} | Private: {profile.is_private}")

for post in profile.get_posts():
    L.download_post(post, target=str(DOWNLOAD_DIR))
```
